```python
import pandas as pd

df = pd.read_csv('Data/10_OneNumSevCatSubgroupsSevObs.csv')
df
```

```python
# Cal total bill
total_bill = df['total_bill'].sum()
print ('total bill = ',total_bill)

total_tips = df['tip'].sum()
print ('total tip = ', total_tips)
```

```python
male = df[df['sex'] == 'Male']
print ('Number of man = ',male.shape[0])

female = df[df['sex'] == 'Female']
print ('Number of man = ',female.shape[0])
```

```python
print ("Prob that a woman book a table = ", (female.shape[0] / (male.shape[0] + female.shape[0]))* 100) 
```

```python
smoke = df[df['smoker'] == 'Yes']
print ('Number of smoker = ',smoke.shape[0])

non_smoke = df[df['smoker'] == 'No']
print ('Number of non smoker = ',non_smoke.shape[0])
```

```python
print ("Prob that a smoker book a table = ", (smoke.shape[0] / (non_smoke.shape[0] + smoke.shape[0]))* 100) 
```

```python
female_smoker = smoke[smoke['sex'] == 'Female']
print ("Prob that a smoker is a women book a table = ", (female_smoker.shape[0] / smoke.shape[0])* 100) 
```
