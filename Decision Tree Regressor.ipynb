## Decision Tree Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/Users/thutranghoa/Code/Data_analysis/Data/Startups_Invest.csv')
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, 4]
```

```python
X.head()
```

```python
y.head()
```

```python
State=pd.get_dummies(X.iloc[:,3],drop_first=True)
```

```python
State.head(10)
```

```python
X.head()
```

```python
X.drop('State',axis=1,inplace=True)
```

```python
X.head()
```

```python
X=pd.concat([X,State],axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeRegressor
decisionregressor=DecisionTreeRegressor()
decisionregressor.fit(X_train,y_train)
```

```python
y_pred=decisionregressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score


LR = LinearRegression()
LR.fit(X_train, y_train)
predictions = LR.predict(X_test)

print ('MSE of LinearRegression= ', mean_squared_error(y_test, predictions))
print ('R2_score of Linear Regression= ', r2_score(y_test, predictions))
```
