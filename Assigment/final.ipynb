```python
'Tỉ lệ bất động sản được bán hoặc thuê theo định kì '

hanoi = [105, -5, 80, 38, 5, 18, 6, 7]
hcm = [110, -20, 58, 78, 15, 10, 42, 6]
x = np.arange(8)

label = [2019, 2020, 2021, 2022, "Q1 2023", "Q2 2023", "Q3 2023", "Q4 2023"]

width = 0.3
fig = plt.figure(figsize=(12, 6))


plt.bar(x- 0.18, hanoi, width, label = 'Hanoi', color = 'teal')
plt.bar(x + 0.18, hcm, width, label = 'Ho Chi Minh', color = 'darkblue')
plt.xticks(x, label)
plt.ylabel('Diện tích cho thuê (nghìn sqm)')
plt.legend()
plt.grid(True, color = "grey", linewidth = "0.7",axis = 'y')
plt.title('Diện tích bất động sản văn phòng được bán và thuê theo định kì')
plt.show()
```

```python
'Tương lai thị trường văn phòng'

'Hanoi'

rent_A = np.array([23, 23.5, 24,24, 24.5, 25.5])
rent_A_cbd = np.array([30, 30.5, 30.5, 31, 32, 34])
rent_B = np.array([13, 14, 14, 14.5, 15, 17])
vacancy_A_CBD = [27, 21, 19, 21, 20, 30]
x = [2019, 2020, 2021, 2022, 2023, 2024]
t = np.arange(len (x))

plt.plot (t,vacancy_A_CBD, color = 'blue', linewidth = "2", label = 'Vacancy - A CBD)')
plt.bar(t, rent_A, color='#621237', label = 'Rent - Grade A')
plt.bar (t, rent_B, bottom= rent_A, color = '#b71f5c', label = 'Rent - Grade B')
plt.bar(t, rent_A_cbd, bottom=rent_B+rent_A , color='#f47e7a', label = 'Rent - Grade A CBD')
plt.xticks(t, x)
plt.legend(loc='upper left')
plt.title ('Hanoi')
plt.show()
```
