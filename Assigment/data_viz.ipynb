```python
import pandas as pd

df1 = pd.read_csv('Data/CEO - Trang tính1.csv')
df2 = pd.read_csv('Data/baoviet - Trang tính1.csv')
df3 = pd.read_csv("Data/Doanxa - Trang tính1.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(data = df1, x = df1['Date'], y = df1['Price'])
plt.xticks(rotation=45)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(data = df2, x = df2['Date'], y = df2['Price'])
plt.xticks(rotation=45)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(data = df3, x = df3['Date'], y = df3['Price'])
plt.xticks(rotation=45)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Data
df=pd.DataFrame({'x_values': df1['Date'], 'y1_values':df1['Price'] , 'y2_values': df2['Price'], 'y3_values': df3['Price'] })
 
# multiple line plots
plt.plot( 'x_values', 'y1_values', data=df, markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4, label = 'CEO')
plt.plot( 'x_values', 'y2_values', data=df, color='olive', linewidth=2, label = "Bao Viet")
plt.plot( 'x_values', 'y3_values', data=df, color='olive', linewidth=2, linestyle='dashed', label="Doanxa")
plt.xticks(rotation=45)
# show legend
plt.legend()

# show graph
plt.show()
```
